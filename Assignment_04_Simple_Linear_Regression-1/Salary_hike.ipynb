## Splitting dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =.25, random_state =1)
```

## Linear Regression Model

```python
from sklearn.linear_model import LinearRegression    ##import linear regression library
model = LinearRegression() 
model.fit(X,Y)
```

```python
y = Y_train
Y_train = model.predict(X_train)
Y_pred = model.predict(X_test)
## predicting Y values for coressponding X values
```

## visualising the train set result

```python
plt.scatter(X_train,Y_train,color ="Red", label="Y_pred")
plt.plot(X_train,Y_train,color ="Green")
plt.scatter(X_train,y,color = "Blue", label ="Actual")
plt.title("Year_Of_Expe Vs Salary_Hike")
plt.xlabel("Year_Of_Expe")
plt.ylabel("Salary_Hike")
plt.legend()
```

## visualising the test set result

```python
plt.scatter(X_test,Y_pred,color ="Red")
plt.plot(X_test,Y_pred,color ="Green", label = "Y_pred")
plt.scatter(X_test,Y_test,color = "Blue", label ="Actual")
plt.title("Delivery_time Vs Sorting_Time")
plt.xlabel("Delivery_Time")
plt.ylabel("Sorting_Time")
plt.legend()
```

##  Accuracy and R^2 value

```python
from sklearn.metrics import r2_score
print("Accuracy is : ", r2_score(Y_test,Y_pred) )
```
